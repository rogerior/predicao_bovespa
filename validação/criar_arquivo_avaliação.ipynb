```python
arquivos = glob.glob('../log/y_test_pred/*')
```

```python
colunas = ['acao', 'tecnica', 'tipo_y', 'teste_quant_0', 'teste_quant_1']

aux = pd.DataFrame(columns=colunas)

for arquivo in arquivos:
    acao = arquivo.split('pred/')[1].split('_')[0]
    tecnica = arquivo.split(acao)[1].split('_')[1]
    tipo_y = arquivo.split(tecnica+'_')[1].split('.')[0]
    
    pd_arquivo = pd.read_csv(arquivo)
    
    quant_0 = np.count_nonzero(pd_arquivo == 0)
    quant_1 = np.count_nonzero(pd_arquivo == 1)
    
    pd_resultado = pd.DataFrame([(acao, tecnica, tipo_y, quant_0, quant_1)], columns=colunas)
    aux = aux.append(pd_resultado)
    
    print(acao)
    #print(tecnica)
    #print(tipo_y)
    #print(quant_0)
    #print(quant_1)
    #break
```

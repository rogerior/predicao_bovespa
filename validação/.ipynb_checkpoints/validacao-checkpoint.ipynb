```python
import pandas as pd
import glob
import sklearn
import numpy as np
import re
from sklearn.externals import joblib

import warnings
warnings.filterwarnings('ignore')
```

```python
dados = pd.read_csv('dados_validacao.csv', sep=';')
dados.head()
```

```python
dados.DATA.unique()
```

```python
len(dados.DATA.unique())
```

```python
acoes = dados.CODNEG.unique()
```

```python
# Função para retornar os dados de uma ação já separados em X e y
def getDadosAcao(acao):
    dados_acao = dados.loc[dados['CODNEG'] == acao]
    
    dados_acao.sort_values('DATA', ascending=True, inplace=True)

    X = dados_acao.copy()
    X.drop(['CODNEG', 'DATA', 'STATUS_POSITIVO'], axis=1, inplace=True)
    y = dados_acao['STATUS_POSITIVO']
    
    return X, y
```

```python
# Função que recebe o y_true e y_pred, aplica os avaliadores de desempenho, faz o comparativo se o desempenho melhorou, e se tiver melhorado chama a função para salvar o novo desempenho no arquivo de controle (csv), posteriormente o modelo desse novo desempenho será salvo pelo joblib (.pkl)
def avaliarDesempenho(y_true, y_pred):
    
    resultados = {}
    # Aplicando avaliadores de desempenho
    resultados['validacao_accuracy'] = [sklearn.metrics.accuracy_score(y_true, y_pred)]
    resultados['validacao_roc_auc_score'] = [sklearn.metrics.roc_auc_score(y_true, y_pred)]
    resultados['validacao_f1_score'] = [sklearn.metrics.f1_score(y_true, y_pred)]
    resultados['validacao_log_loss'] = [sklearn.metrics.log_loss(y_true, y_pred)]
    resultados['validacao_precision'] = [sklearn.metrics.precision_score(y_true, y_pred)]
    resultados['validacao_recall'] = [sklearn.metrics.recall_score(y_true, y_pred)]
    
    
    pd_resultados = pd.DataFrame(data=resultados)
    

    return pd_resultados
```

para cada ação
get dados da ação
carregar cada modelo da ação (for)
para cada modelo
realizar predição
aplicar avaliadores de desempenho
salvar resultado

```python
def validacao(acao):
    print( 'Ação: ' + acao)
    resultados = []    
    
    X, y = getDadosAcao(acao)
    
    valid_true_quant_0 = np.count_nonzero(y == 0)
    valid_true_quant_1 = np.count_nonzero(y == 1)    
    
    valid_tamanho_base = len(X)

    modelos = glob.glob('../modelos/'+acao+'_*')
    
    # Se existir modelo salvo
    if len(modelos) > 0:

        for modelo in modelos:

            tecnica = modelo.split('_')[1].split('.')[0]
            #print( 'Ação: ' + acao + ', técnica: ' + tecnica )

            model = joblib.load(modelo)

            y_pred = model.predict(X)
            
            # Salvando y_test e y_pred para futuras verificações caso necessário
            y.to_csv('log/' + acao + '_' + tecnica + '_y_true.csv', index=False, header=True)
            pd.DataFrame(y_pred, columns=['y_pred']).to_csv('log/' + acao + '_' + tecnica + '_y_pred.csv', index=False)


            valid_pred_quant_0 = np.count_nonzero(y_pred == 0)
            valid_pred_quant_1 = np.count_nonzero(y_pred == 1)        

            resultados_avaliadores = avaliarDesempenho(y_true=y, y_pred=y_pred)

            resultado = {}
            resultado['acao'] = [acao]
            resultado['tecnica'] = [tecnica]
            resultado['validacao_y_true_quant_0'] = [valid_true_quant_0]
            resultado['validacao_y_true_quant_1'] = [valid_true_quant_1]
            resultado['validacao_tamanho_base'] = [valid_tamanho_base]
            resultado['validacao_y_pred_quant_0'] = [valid_pred_quant_0]
            resultado['validacao_y_pred_quant_1'] = [valid_pred_quant_1]

            pd_resultado = pd.DataFrame(data=resultado)

            resultado = pd.concat([pd_resultado, resultados_avaliadores], axis=1)

            if len(resultados) == 0:
                resultados = resultado
            else:
                resultados = pd.concat([resultados, resultado])
                
    # Se não existir modelo salvo
    else:
        return None
            
    return resultados
```

```python
resultados = []
for acao in acoes:
    
    resultado = validacao(acao)
    
    if resultado is not None:

        if len(resultados) == 0:
            resultados = resultado
        else:
            resultados = pd.concat([resultados, resultado])
    
```

```python
resultados.to_csv('resultados_validacao.csv', sep=';', index=False)
```
